v  Load Train Data  v

```python
import requests
import pandas as pd
from io import BytesIO

def loadData(link):
  req = requests.get(link)
  content = BytesIO(req.content)
  df = pd.read_csv(content)
  return df

food_data = loadData("https://raw.githubusercontent.com/ezramuskat/Ingredient-Substitution-Capstone/main/data_preparation/classification_dataset/common_ingredients.csv")
food_data
```

v  Filter Ingredients List  v

```python
from filtering_model import FilteringModel

fmodel = FilteringModel(food_data)
fmodel.train_model(epochs=20,batch_size=33,val_split=0.0)
filtered_data = fmodel.filter(loaded_ingredients[:3000])
```
