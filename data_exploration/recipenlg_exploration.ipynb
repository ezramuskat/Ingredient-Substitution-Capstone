Now check out some collation stuff

```python
import re
def strip_units(ingredient: str) -> str:
    # Define common units of measurement
    units = [
        r"\b(?:cup[s]?|tsp[s]?|tbsp[s].?|ounce[s]?|oz|pound[s]?|lb[s]?|gram[s]?|g|kg|liter[s]?|l|ml|pinch|dash|quart[s]?|qt|pint[s]?|pt|gallon[s]?|gal?|c.?|pkg.)\b",
        r"\b(?:dozen|clove[s]?|stick[s]?|can[s]?|bottle[s]?|packet[s]?|slice[s]?|piece[s]?)\b",
        r"\d+\/\d+",  # Matches fractions like 1/2, 3/4
        r"\d+\s*\(?\d*\.?\d*\)?",  # Matches whole numbers, decimals, and mixed fractions (e.g., 1, 1.5, (1/2))
    ]

    # Remove units and extra spaces
    cleaned_ingredient = re.sub("|".join(units), "", ingredient, flags=re.IGNORECASE).strip()

    return cleaned_ingredient
```

```python
ingredients = df['ingredients'].to_list()
ingredients = list(map(strip_units, ingredients))
ingredients
```
